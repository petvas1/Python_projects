# Basic 2D Plotting with Matplotlib

## Heatmaps

```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

Loading the data (2D matrix)

```python
df = pd.read_csv("temperature_data.csv", index_col=0)
```

```python
df
```

```python
df = df.loc[::-1]  # Reverse the DataFrame's rows to match the correct latitude order
array = df.to_numpy()
```

Basic plot

```python
plt.imshow(array)
plt.show()
```

Adding plot elements:

```python
lon = df.columns.astype(float)   # x-axis values
lat = df.index                   # y-axis values
from matplotlib.ticker import StrMethodFormatter  # for formatting ticks
```

```python
df_msl = pd.read_csv("msl_data.csv", index_col=0)
df_msl = df_msl.loc[::-1]
```

```python
array_msl = df_msl.to_numpy()
```

```python
from scipy import ndimage
array_msl = ndimage.zoom(array_msl, 100)
lon = ndimage.zoom(lon, 100)
lat = ndimage.zoom(lat, 100)
```

```python
# ---- Load CSV and make quiver plot ----
df = pd.read_csv("wind_data.csv")

# Reshape data (assuming grid structure)
lat_unique = np.sort(df.latitude.unique())
lon_unique = np.sort(df.longitude.unique())
lon_grid, lat_grid = np.meshgrid(lon_unique, lat_unique)

# Reshape wind components
u_grid = df.u.values.reshape(len(lat_unique), len(lon_unique))
v_grid = df.v.values.reshape(len(lat_unique), len(lon_unique))
```

```python
fig, ax = plt.subplots()
plt.imshow(array, cmap='turbo', interpolation='spline16', extent=(min(lon), max(lon), min(lat), max(lat)))
cb = plt.colorbar(ticks=np.arange(-10, 10, 2))  
cb.set_label('Temp [°C]', fontsize=10, rotation='horizontal', labelpad=-20, y=1.05)
ax.xaxis.set_major_formatter(StrMethodFormatter('{x:.0f}° E'))
ax.yaxis.set_major_formatter(StrMethodFormatter('{x:.0f}° N'))
plt.xlabel('Longitude')
plt.ylabel('Latitude')
# plt.title('Temperature heatmap')

cs = plt.contour(lon, lat, array_msl, colors='k', linewidths=0.7)
plt.clabel(cs, fontsize=8, colors='k', inline=True)
# plt.savefig("temperature_heatmap.png", dpi=300)  # Uncomment to save the figure

plt.quiver(lon_grid, lat_grid, u_grid, v_grid, scale=100, color="blue")
plt.show()
```
